```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train\\train.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data = data.dropna(axis=1, how= 'all')
```

```python
X = data.loc[:, data.columns != 'row_id']
X = X.loc[:, X.columns != 'rating_num']
X.describe()
```

```python
X['scout_id'].dtype
```

```python
X[X.columns[X.dtypes!='object']]
```

```python
X
```

```python
from sklearn import impute
imp = impute.SimpleImputer(missing_values=np.nan, strategy = 'mean')
imp.fit(X[X.columns[X.dtypes!='object']])
X[X.columns[X.dtypes!='object']] = imp.transform(X[X.columns[X.dtypes!='object']]) 
```

```python
X.columns.isna()
```

## Label Encoder

```python
for i in X.columns[X.dtypes=='object']:
    print(i)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

for i in X.columns[X.dtypes=='object']:
    
    le.fit(X[i])
    le.classes_
    X[i] = le.transform(X[i])
```

```python
X
```

```python
Y = data.iloc[:, data.columns=='rating_num'].values
Y
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
stdsc = StandardScaler()
X = stdsc.fit_transform(X)
print(X) 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.15, random_state = 1)
```

```python
from xgboost import XGBRegressor
```

```python

# fit model no training data
model = XGBRegressor()
model.fit(X_train, Y_train)
```

```python
print(model)
```

```python
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
from sklearn.metrics import r2_score, mean_squared_error

r2_score(Y_test, y_pred)
```
